## ZCR

### Fastest talker

```python
# Load the audio file
file_path = 'audio1.m4a'
y, sr = librosa.load(file_path, sr=None)

# Define the time range (25s to 50s)
start_time = 25.0
end_time = 50.0

# Extract segment
start_sample = int(start_time * sr)
end_sample = int(end_time * sr)
y_segment = y[start_sample:end_sample]

# Time vector for selected segment
duration_segment = len(y_segment) / sr
time_segment = np.linspace(0, duration_segment, len(y_segment))

# Calculate Zero Crossings
zero_crossings = np.sum(librosa.zero_crossings(y_segment, pad=False))
zero_crossing_rate = zero_crossings / (duration_segment*4)

print(f"Sample rate: {sr} Hz")
print(f"Segment duration: {duration_segment:.2f} seconds")
print(f"Total Zero Crossings in segment: {zero_crossings}")
print(f"Zero Crossing Rate (per second) in segment: {zero_crossing_rate:.2f}")

# Plot the waveform
plt.figure(figsize=(12, 6))
plt.plot(time_segment, y_segment, color="#80BCCE")
plt.title('Fastes talker Steve Woodmore (Guiness Record 1990)')

# Display Zero Crossing Rate on plot
plt.text(25, max(y_segment) * 0.96, f'ZCR: {zero_crossing_rate:.3f}',
         fontsize=11, color='black', ha='right', va='top',
         bbox=dict(facecolor='white', alpha=0.7, edgecolor='black', boxstyle='round,pad=0.5'))

# Labels and grid
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.ylim(-1.1,1.1)
plt.savefig("/content/drive/MyDrive/wszystkie_rzeczy_z_colaba/wykresy/fastest_talker.png")
plt.show()
```

## RMS Amplitude

### We will rock you

```python
# Load the audio file
file_path = 'audio5.m4a'
y, sr = librosa.load(file_path, sr=None)

# Define the time range (25s to 50s)
start_time = 0.0
end_time = 24.0

# Extract segment
start_sample = int(start_time * sr)
end_sample = int(end_time * sr)
y_segment = y[start_sample:end_sample]

# Time vector for selected segment
duration_segment = len(y_segment) / sr
time_segment = np.linspace(0, duration_segment, len(y_segment))

# --- Instead of Zero Crossing, compute Mean RMS Amplitude ---
rms = librosa.feature.rms(y=y_segment)[0]
mean_rms = np.mean(rms)

print(f"Sample rate: {sr} Hz")
print(f"Segment duration: {duration_segment:.2f} seconds")
print(f"Mean RMS Amplitude in segment: {mean_rms:.5f}")

# --- Plot the waveform ---
plt.figure(figsize=(12, 6))
plt.plot(time_segment, y_segment, color="#80BCCE")
plt.title('We Will Rock You! - Queen')

# Display Mean RMS on plot
plt.text(25-0.28, max(y_segment) * 1.26, f'Mean RMS Amplitude: {mean_rms:.5f}',
         fontsize=11, color='black', ha='right', va='top',
         bbox=dict(facecolor='white', alpha=0.7, edgecolor='black', boxstyle='round,pad=0.5'))

# Labels and grid
plt.xlim(0,25)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.ylim(-1.1,1.1)
plt.savefig("/content/drive/MyDrive/wszystkie_rzeczy_z_colaba/wykresy/queen.png")
plt.show()
```
